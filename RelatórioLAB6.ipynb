### 2) **Velocidade ap√≥s o quique**

$\qquad$Para encontrarmos a velocidade ap√≥s o quique, iremos utilizar a equa√ß√£o de Torricelli. $\;$ A equa√ß√£o de Torricilli √© definidade por:
\begin{align}\\
        v^2 = v^2_0 - 2\cdot a\cdotŒîS 
    \end{align}

$\qquad$onde,  
\begin{align}\\
        v&: \; velocidade \;  final \; (m/s)\\
        v_0&: \; velocidade \;  inicial \; (m/s)\\
        \Delta S &: \; espa√ßo \;  percorrido \; (m)\\
        a&: \; acelera√ß√£o \;  da \;  gravidade \; (m/s^2) \; ‚Üí \;  a=g=9,81 \; m/s^2\\\\
    \end{align}
$\qquad$para $\;$ $v^2=0$, $\;$ temos que \\
\begin{align}\\\\
        -v_0^2 &=  - 2\cdot a\cdotŒîS \quad \cdot(-1)\\ 
        \\ v_0^2 &=   2\cdot a\cdotŒîS\\ 
        \\ |v_0| &=   \sqrt{2\cdot a\cdotŒîS}\\
        \\ v_0 &=   ¬±\sqrt{2\cdot a\cdotŒîS}\\
        \\ v_0 &=   \sqrt{2\cdot a\cdotŒîS}\\\\      
    \end{align}

### 3) **Resolvendo o sistema h√≠brido de equa√ß√µes diferenciais**

$\qquad$Atrav√©s da equa√ß√£o horaria do espa√ßo, podemos montar o sistema de equa√ß√µes diferenciais ordin√°rias que vai nos auxiliar na solu√ß√£o 

$\qquad$pelos m√©todos de Euler e RK2. O sistema pode ser definido da seguinte forma:
\begin{align}\\
        y = y_0 + v_0 \cdot t - \frac{1}{2} \cdot g \cdot t^2
        \\\text{ } 
    \end{align}

$\qquad$onde,

\begin{align}\\
        \frac{dy}{dt}&=v_0-g \cdot t\\
        \\\frac{d^2y}{d^2t}&=-g
        \\\text{ }  
    \end{align}


$\qquad$para deixarmos o sistema com ordem equivalente, com uma vari√°vel auxiliar, $\; $tomamos $v \; = \;v_0 \:-g  \cdot t$ ,$\;$ ent√£o

\begin{align}\\
        y'=\frac{dy}{dt}&=v_0-g \cdot t = v\\
        \\ &\boxed{y'=v}
        \\\text{ }  
    \end{align}

$\qquad$e

\begin{align}\\
        v &= v_0 -g \cdot t\\
        \\ &\boxed{v'=-g}
        \\\text{ } 
    \end{align}

$\qquad$portanto, $\;$ o sistema EDO tem como problema de valor inicial (PVI) $\; y(t) \; e \; v(t) \;$, partindo de $\;t=1,530 \;$

\begin{align}\\\\
        \left\{
        \begin{array}{cl}
        y'&=v\\
        v'&=-g
        \\y(1,530 ) &= 1,3670 \qquad v(1,530 )=0
        \end{array}
        \right.
        \\\text{ } 
    \end{align}


  

* **M√©todo de Euler**

  Resolvendo o sistema utilizando o algoritmo do m√©todo de Euler para os passos de $\; h(0.1\; e \;0.01) \; $, obtemos os seguintes gr√°ficos.

**Algoritmo do m√©todo de Euler com passo h=0.1**

```python
#m√©todo de Euler (h=0.1)
import numpy as np
import matplotlib.pyplot as plt

h = 0.1

def metodo_de_euler(h):

  #dados
  t = np.arange(1.530,4.168,h)  #tempo inicial e final das alturas m√°ximas da bola
  n = np.size(t)
  v = np.zeros(n) #defini√ß√£o do vetor velocidade
  y = np.zeros(n) #defini√ß√£o do vetor altura
  g = -9.81 
  y[0] = 1.3670 #altura m√°xima inicial
  v[0] = 0 #velocidade inicial da bola
  
  k = 0.8758 #constante de amortecimento

  #algoritmo do m√©todo de Euler
  for i in range(1,n):

      y[i] = y[i-1]+(h*v[i-1])
      v[i] = v[i-1]+ (h*g)

      #condi√ß√£o discreta
      if y[i] <= 0.02:

        v[i] = abs(k*v[i-1])

  #plot
  plt.figure(figsize=(15,7))
  plt.plot(t, y, color='blue' ,label="Quique da bola")
  y1 = np.array([1.3670, 0.9850,0.7555, 0.6016, 0.5089, 0.4433])
  t1 = np.array([1.530, 2.268, 2.864, 3.402, 3.802, 4.168])
  plt.plot([t1], [y1], 'bs', markersize=10, color='red') #PONTOS
  plt.plot(1.3670, 1.530, 'bs', markersize=10, color='red', label="Amostras")
  plt.ylim(-0.2,1.5)
  plt.grid(True)
  plt.xlabel("Tempo(s)", fontsize =20)
  plt.ylabel("Altura(m)", fontsize =20)
  plt.title('Trajet√≥ria da bola', fontsize=20)
  plt.legend(loc="upper right")
  plt.show()
  print("\n\n")
  plt.figure(figsize=(15,7))
  plt.title('Quique da bola', fontsize=20)
  plt.grid(True)
  plt.xlabel("Tempo(s)", fontsize =20)
  plt.ylabel("Velocidade(m/s)", fontsize =20)
  plt.plot(t, v, color='y' ,label="velocidade da bola")
  plt.legend(loc="upper right")
  plt.show()

#chamada da fun√ß√£o
metodo_de_euler(h)
```

**Algoritmo do m√©todo de Euler com o passo de h=0.01**

```python
#m√©todo de Euler (h=0.01)
h = 0.01

#chamada da fun√ß√£o
metodo_de_euler(h)
```

* **M√©todo de Runge Kutta ordem 2**

  Resolvendo o sistema utilizando o algoritmo do m√©todo de Runge Kutta de ordem 2 (RK2) para os espa√ßamentos de  $\;h(0.1\;e\;0.01)$, obtemos os seguintes gr√°ficos.

**Algoritmo do m√©todo de Runge Kutta com o espa√ßamento de h=0.1**

```python
#m√©todo de Runge Kutta (h=0.1)

k = 0.8758 #constante de amortecimento
h = 0.1 #passos
a = 1.530  #x0
b = 4.168  #xn
x_inicial = 1.530 #valor x inicial da PVI(TEMPO 't')
y_inicial = 1.3670 #valor Y inicial da PVI
v_inicial = 0 #valor v inicial da PVI

#fun√ß√£o velocidade
def diff_v(v):

    z = v

    return z

#fun√ß√£o acelera√ß√£o
def diff_g(g):

    a = -g

    return a

#fun√ß√£o de RK2
def RK2(x_inicial,y_inicial,v_inicial,a,b,h,k):
    
  #dados
  n = int((b-a)/h)+1 #numero de itera√ß√µes
  x = np.linspace(a,b,n) #defini√ß√£o do vetor x
  x[0] = x_inicial #x inicial
  y = np.zeros(n) #defini√ß√£o do y
  v = np.zeros(n) #defini√ß√£o de v
  y[0] = y_inicial #y inicial
  v[0] = v_inicial #y inicial
  g = 9.81
    
  #algoritmo do m√©todo de RK2
  for i in range(1,n):

      #velocidade
      k1 = diff_v(v[i-1])
      k2 = diff_v(v[i-1]+h*k1)
      y[i] = y[i-1]+(h/2)*(k1+k2)

      #acelera√ß√£o
      k1v = diff_g(g)
      k2v = diff_g(g+h*k1)
      v[i] = v[i-1]+(h/2)*(k1v+k2v)

      #condi√ß√£o discreta
      if y[i] < 0.2:

        v[i] = -k*v[i-1] 

  #plot do gr√°fico
  plt.figure(figsize=(15,7))
  plt.plot(x, y, color='blue' ,label="Quique da bola")
  y1 = np.array([1.3670, 0.9850,0.7555, 0.6016, 0.5089, 0.4433])
  x1 = np.array([1.530, 2.268, 2.864, 3.402, 3.802, 4.168])
  plt.plot([x1], [y1], 'bs', markersize=10, color='red') #PONTOS
  plt.plot(1.3670, 1.530, 'bs', markersize=10, color='red', label="Amostras")
  plt.ylim(-0.2,1.5)
  plt.grid(True)
  plt.xlabel("Tempo(s)", fontsize =20)
  plt.ylabel("Altura(m)", fontsize =20)
  plt.title('Quique da bola', fontsize=20)
  plt.legend(loc="upper right")
  plt.show()
  print("\n\n")
  plt.figure(figsize=(15,7))
  plt.title('Velocidade da bola', fontsize=20)
  plt.grid(True)
  plt.xlabel("Tempo(s)", fontsize =20)
  plt.ylabel("Velocidade(m/s)", fontsize =20)
  plt.plot(x, v, color='y' ,label="Quique da bola")
  plt.legend(loc="upper right")
  plt.show()

#chamada da fun√ß√£o
RK2(x_inicial,y_inicial,v_inicial,a,b,h,k)
```

**Algoritmo do m√©todo de Runge Kutta com o espa√ßamento de h=0.01**

```python
#m√©todo de Runge Kutta (h=0.01)
h = 0.01 #passo

#chamada da fun√ß√£o
RK2(x_inicial,y_inicial,v_inicial,a,b,h,k)
```
